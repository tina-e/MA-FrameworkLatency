```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
files = os.listdir("../data/test_windup_py")
files = [x for x in files if 'csv' in x]
files
```

```python
dataframes = []
for file in files:
    dataframes.append(pd.read_csv(f"../data/test_windup_py/{file}"))
data = pd.concat(dataframes)
data
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.boxplot(data=data, x='program', y='ete', hue='fw_running').set_title('end-to-end latencies with/without running fw-tester')
sns.stripplot(data=data, x='program', y='ete', hue='fw_running', dodge=True, color='black')
ax = plt.gca()
ax.set_ylim(20000, 60000)
for i in range(0, 20):
    plt.axhline(i*16.7*1000, color='red')
```

```python
sns.boxplot(data=data, x='program', y='fw', hue='program_fwrunning', dodge=True).set_title('')
sns.stripplot(data=data, x='program', y='fw', hue='program_fwrunning', dodge=True, color='black')
```
