```python
import pandas as pd
import numpy as np
import math
import os
```

```python
FRAMEWORKS = ["Direct3D","FLTK","GLEW_SDL","GLUT","GTK","Java2D","JavaSwing","pygame","pyglet","pyqt5","pyqt6","Qt5","Qt6","SDL2_OpenGL","SDL2_OpenGLES2","SDL2_OpenGLES","SDL2_Direct3D","SDL2_Software","tkinter","win32","wxpython"]
PATH = "../data/vsync_tester_0605_60hz/"
```

```python
files = os.listdir(PATH)
file_data = []
for file in files:
    data_in = pd.read_csv(f"{PATH}/{file}")
    file_data.append(data_in)
    
data = pd.concat(file_data)
data
```

```python
for framework in FRAMEWORKS:
    data_fw = data[data['framework'] == framework]
    vsync_counter= len(data_fw[data_fw['ete_upper'] > data_fw['ete_lower']])

    print(f'{framework}: vsync off detected {vsync_counter} times')
```
