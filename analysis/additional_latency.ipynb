# Is there any additional latency caused by framework tester?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
path = "../data/first_measurements_pilot/"  # todo: adjust path! (prob. only one file)
files = os.listdir(path)
```

```python
dataframes = []
for file in files:
    dataframes.append(pd.read_csv(f"{path}/{file}"))
data = pd.concat(dataframes)
```

```python
sns.boxplot(data=data, x='program', y='ete', hue='fw_running').set_title('end-to-end latencies with/without running fw-tester')
sns.stripplot(data=data, x='program', y='ete', hue='fw_running', dodge=True)
```
