```python
import random
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
fw_program = random.choices(range(4000, 6000), k=200)
ete_program = random.choices(range(6000, 9000), k=200)

data_program = pd.DataFrame(columns=['ete', 'fw', 'diff'])
data_program['ete'] = ete_program
data_program['fw'] = fw_program
data_program['diff'] = data_program['ete'] - data_program['fw']
data_program['program_running'] = 'program running' 

data_program
```

```python
ete_none = random.choices(range(6000, 9000), k=200)
fw_none = random.choices(range(4000, 6000), k=200)

data_none = pd.DataFrame(columns=['ete', 'fw', 'diff'])
data_none['ete'] = ete_none
data_none['fw'] = fw_none
data_none['diff'] = data_none['ete'] - data_none['fw']
data_none['program_running'] = 'program not running'

data_none
```

```python
data = pd.concat([data_program, data_none])
data
```

```python
plot = sns.stripplot(data=data, x='program_running', y='ete', order=['program running', 'program not running'], palette=['grey', 'darkred'], size=1.5)
plot.axhline(data[data['program_running'] == 'program running']['ete'].median(), color='darkred', alpha=1, linestyle='--')
quant_25 = np.percentile(data_none['ete'], 25)
quant_75 = np.percentile(data_none['ete'], 75)
plot.axhspan(quant_25, quant_75, facecolor='r', alpha=0.25, label="test")
```
