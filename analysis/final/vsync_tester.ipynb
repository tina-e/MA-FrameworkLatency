```python
import pandas as pd
import numpy as np
import math
import os
```

```python
PATH = "../../data/finals_all_0106/"
```

```python
files = os.listdir(PATH)
file_data = []
for file in files:
    data_in = pd.read_csv(f"{PATH}/{file}")
    file_data.append(data_in)
    
data = pd.concat(file_data)
data
```

```python
FRAMEWORKS = data['framework'].unique()
```

```python
data = data[data['complexity'] == 'default']
data = data[data['fullscreen'] == 'fullscreen']
# data = data[data['program'] != 'getdbits']
# data = data[data['program'] != 'bitblt']
# data = data[data['program'] != 'none']
```

```python
vsync_off_frameworks = []
for framework in FRAMEWORKS:
    data_fw = data[data['framework'] == framework]
    vsync_counter= len(data_fw[data_fw['ete_upper'] > data_fw['ete_lower']])

    print(f'{framework}: vsync off detected {vsync_counter} times')
    if vsync_counter > 0:
        vsync_off_frameworks.append(framework)
print(vsync_off_frameworks)
```
