```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.transforms as transforms
from matplotlib.legend_handler import HandlerTuple
import seaborn as sns
import math
import os
```

```python
PATH = "../data/2802_all_and_fullscreens"
files = os.listdir(PATH)
files = [x for x in files if 'csv' in x]
```

```python
file_data = []
for file in files:
    data_in = pd.read_csv(f"{PATH}/{file}")
    file_data.append(data_in)
    
data = pd.concat(file_data)
```

```python
data.loc[data.fw_running == False, 'program'] = 'none'
data.loc[data.program == 'windup_fullscreen', 'program'] = 'windup'
```

```python
data = data.drop('fw_running', axis=1)
data = data.drop('program_fwrunning', axis=1)
```
